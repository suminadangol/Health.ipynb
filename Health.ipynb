```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# LOAD DATA SET

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

# LINE CHART


```python
sns.lineplot(data = df , x= 'age' , y = 'BMI' , marker = '*')
plt.title ("Age Vs BMI")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

Here, BMI is mostly high in between the age to 30 ,after the age 30 it's in the decreasing form and after 50 there isn't gradual increment in BMI but can see that compared to 30-50 age there's high BMI.

# BAR PLOT

```python
sns.barplot(x='sex',y='weightKg',data=df)
```

The weight of both males and females is approximately the same.

# PIE CHART

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

The female population is outnumbering the male population.

# BOX PLOT

```python
sns.boxplot(x='sex',y='age',hue='BMI',data=df)
```

As we can see ,BMI of males appears to be more favorable compared to females.

# HISTOGRAM

```python
sns.displot(df['age'], kind="hist", kde=True)
```

 Age 20 is the most prevalent.

# PLOT

```python
sns.scatterplot(x='age', y='DiaBP',hue='sex',data=df)
plt.show()
```

Male appear to be more dominant.
